# Singly Linked Lists

```python
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_begining(self, data):
        node = Node(data, self.head)
        self.head = node
    
    def print(self):
        if self.head is None:
            print("Linked List is empty")
            return
        
        itr = self.head
        
        while itr:
            print(itr.data)
            itr = itr.next

    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
        
        itr.next = Node(data, None)
    
    def insert_values(self, data_list):
#         self.head = None                : if u want to empty list first
        for data in data_list:
            self.insert_at_end(data)
        
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        
        return count        

    def remove_at(self, index):
        if index<0 or index>=self.get_length():
            raise Exception("Invalid Index")
        
        if index==0:
            self.head = self.head.next
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == index-1:
                itr.next = itr.next.next
                break
                
            itr = itr.next
            count += 1
        
    def insert_at(self, index, data):
        if index<0 or index>=self.get_length():
            raise Exception("Invalid Index")
        
        if index==0:
            self.insert_at_begining(data)
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == index-1:
                node = Node(data, itr.next)
                itr.next = node
                break
            itr = itr.next
            count += 1
    
    def insert_after_value(self, data_after, data_to_insert):
        count = 0
        itr = self.head
        while itr:
            if itr.data == data_after:
                node = Node(data_to_insert, itr.next)
                itr.next = node
                break
            itr = itr.next
            count += 1
            if count == self.get_length():
                print("Invalid data after")
     
    def remove_by_value(self, data):
        count = 0
        itr = self.head
        itr_prev = self.head
        while itr:
            if itr.data == data:
                if itr == self.head:
                    self.head = itr.next
                else:
                    itr_prev.next = itr.next
            
            itr_prev = itr
            itr = itr.next
            count += 1
        if count == self.get_length():
            print("Data not found")
            

if __name__ == '__main__':
    ll = LinkedList()

#     ll.insert_at_begining(5)
#     ll.insert_at_begining(23)

#     ll.insert_at_end(225)
#     ll.insert_at_end(22)
#     ll.insert_at_end(2)
    
#     ll.insert_values(["hii", "yoo", "Hola", "Adois!!"])
#     ll.print()
    
#     print("Length: ",ll.get_length())
    
#     ll.remove_at(2)
#     ll.print()
# #     ll.remove_at(-1)

#     ll.insert_at(2,"holaa")
#     ll.print()
    
#     ll.insert_after_value("Adois!!", "Sarvesh")
#     ll.print()
    
#     ll.remove_by_value("hii")
#     ll.print()
```

# Doubly Linked Lists

```python
class Node:
    def __init__(self, data = None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev

class DoublyLinkedList:
    def __init__(self):
        self.head = None     # points to the beginner node of the list
        self.tail = None     # points to the latest node added to the list
        self.count = 0

    def print(self):
        if self.head is None:
            print("Linked List is empty")
            return
        
        itr = self.head
        while itr:
            print(itr.data)
            itr = itr.next
            
    def append(self, data):
        node = Node(data, None, None)
        if self.head == None:
            self.head = node
            self.tail = self.head
        else:
            node.prev = self.tail
            self.tail.next = node
            self.tail = node
            self.count += 1
    
    def delete_by_value(self, data):
        current = self.head
        node_deleted = False
        
        if current is None:             # List is empty
            node_deleted = False
            print("List is empty")
        
        elif current.data == data:        # Item to be deleted is found at the start of list.
            self.head = current.next
            self.head.prev = None
            node_deleted = True
        
        elif self.tail.data == data:     # Item to be deleted is found at the end of list.
            self.tail = self.tail.prev
            self.tail.next = None
            node_deleted = True
        
        else:
            while current:
                if current.data == data:
                    current.prev.next = current.next
                    current.next.prev = current.prev
                    node_deleted = True
                current = current.next
        
        if node_deleted:
            self.count -= 1
        else:
            print("Data not found")
            
    def contain(self, data):
        iter = self.head
        while iter:
            if iter.data == data:
                return True
            iter = iter.next
        return False    
    
    

dl = DoublyLinkedList()
dl.append(5)
dl.append(123)
dl.append(23)
dl.append(20)
dl.append(67)
dl.print()

print()

dl.delete_by_value(1)
dl.print()

if dl.contain(123):
    print("present")
else:
    print("Not present")
```

# Circular Lists

```python
class Node():
    
    def __init__(self, data = None, next = None):
        self.data = data
        self.next = next
        
class CircularLinkedList():
    
    def __init__(self, head = None, end = None):
        self.head = head
        self.end = end
    
    def traverse(self):
        
        # define the first node
        curr_node = self.head
        
        # as long as there is a next node keep going
        while curr_node.next:
            
            # print the data
            print(curr_node.data)
            curr_node = curr_node.next
            
            # once we get back to the head, end the loop
            if curr_node == self.head:
                break
        
    def insert_end(self, data):
        
        new_node = Node(data)
        
        # if the list is empty
        if self.head == None:
            self.head = new_node
            self.head.next = new_node
            self.end = new_node
            return
        
        # if the list is not empty
        if self.end != None:
            self.end.next = new_node
            new_node.next = self.head
            self.end = new_node
            return
    
    def insert_beg(self, data):
        
        new_node = Node(data)
        new_node.next = self.head
        curr_node = self.head
        
        # handle the empty list case
        if curr_node == None:
            self.head = new_node
            self.end = new_node
            self.head.next = new_node
            return
        
        # handle the non empty liost case
        if self.end != None:
            self.end.next = new_node
            new_node.next = self.head
            self.head = new_node
            return

    def insert_mid(self, ref_node, data):

        # handle empty node case
        if ref_node == None:
            raise ValueError("The reference can't be none")

        # if it's the end one, use the existing insert_end method
        if ref_node == self.end:
            self.insert_end(data)
        else:
            # otherwise it's a true mid
            new_node = Node(data)
            ref_next = ref_node.next
            ref_node.next = new_node
            new_node.next = ref_next
            
    def delete_beg(self):
        
        if self.head != None:
            # grab the node that comes after the head
            aft_head = self.head.next

            # have the last node point to the node after the head
            self.end.next = aft_head

            # set the new head
            self.head = aft_head
        else:
            raise ValueError("The list is empty")
        
    def delete_end(self):
        
        if self.head != None:
            
            # grab the head
            curr_node = self.head
            
            # traverse untill end
            while curr_node.next.next != self.head:
                curr_node = curr_node.next
            
            # reassign the tail
            self.end = curr_node
            
            # have the new last node point to the head
            curr_node.next = self.head
        else:
            raise ValueError("The list is empty")
    
    def delete_mid(self, position):
        
        # if position is 0 then delete first
        if position == 0:
            self.delete_beg()
            return
        
        # if position is the size of the list minus one then delete end
        if position == self.list_size()-1:
            self.delete_end()
            return
        
        if position > self.list_size() - 1:
            print("Invalid Position")
            return
        
        # grab the node after the head
        curr_node = self.head.next
        
        # initialise the count
        count = 1
        
        # traverse till you reach the position
        while count < position-1:
            count += 1
            curr_node = curr_node.next
            
        # have it point to the one after the deleted one
        curr_node.next = curr_node.next.next
        
    def list_size(self):
        
        # grab the head
        curr_node = self.head
        count = 0
        
        # traverse until you reach the head again
        while curr_node.next:
            
            count += 1
            curr_node = curr_node.next

            # once we get back to the head end the loop
            if curr_node == self.head:
                break
        
        return count
        
        
            
cl = CircularLinkedList()
cl.insert_end(50)
cl.insert_end(60)
cl.insert_end(70)
cl.insert_end(860)

cl.insert_beg(90)
cl.insert_beg(100)
cl.insert_beg(54)

print('After Insertion')
print('-'*20)
cl.traverse()

fnode = cl.head            # grab the first node
cl.insert_mid(fnode, 20)

print('After Mid Insertion')
print('-'*20)
cl.traverse()

cl.delete_beg()
cl.delete_end()
cl.delete_mid(4)          # zero base position is deleted; for one base: set count = 2

print('After Deletion')
print('-'*20)
cl.traverse()
```
